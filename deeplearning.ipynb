```python
import tensorflow
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
dados = keras.datasets.fashion_mnist
((imagens_treino, identificacoes_treino),(imagens_teste, identificacoes_teste)) = dados.load_data()
```

```python
print(len(imagens_treino))
print(imagens_treino.shape)
```

```python
print(len(imagens_teste))
print(imagens_teste.shape)
```

```python
print(identificacoes_treino.shape)
print(identificacoes_teste.shape)
```

```python
plt.imshow(imagens_treino[0], cmap='Greys')
plt.title(identificacoes_treino[0])
plt.show()
```

```python
print(identificacoes_treino.min())
print(identificacoes_treino.max())
```

```python
total_de_classificacoes = len(set(identificacoes_treino))
print(total_de_classificacoes)
```

```python
for imagem in range(total_de_classificacoes):
    plt.imshow(imagens_treino[imagem], cmap='Greys')
    plt.title(identificacoes_treino[imagem])
    plt.show()
```

```python
nomes_de_classificacoes = [
    'Camisa','Cala','Pullover','Vestido','Casaco',
    'Sandlia','Camisa','Tnis','Bolsa','Bota'
]
```

```python
for imagem in range(total_de_classificacoes):
    plt.subplot(2 , 5, imagem+1)
    plt.imshow(imagens_treino[imagem], cmap='Greys')
    plt.title(nomes_de_classificacoes[identificacoes_treino[imagem]])
    plt.show()
```
